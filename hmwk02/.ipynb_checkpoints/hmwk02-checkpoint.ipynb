### [20 points] Problem 1

In a box there are 99 fair coins and 1 double-headed coin.  You grab one coin at random.  

Work the following problems by hand (but don't be afraid to simulate to check your result!). 


**Part A**: You toss your coin one time.  What is the probability that you toss a Heads? 

We have 99 fair coins, with a 50-50 chance to flip heads. 

Likewise, we have one coin that has a 1/1 chance to flip heads.

Since the only variant in this question is whether or not we choose a biased coin or a fair coin, we can add the probabilities of eachother together.

Therfore, the probability that you flip heads, given that you have a biased coin is:

$$P(H) = \frac{99}{100}x\frac{1}{2} + \frac{1}{100}x\frac{1}{1}$$

$$ P(H) = \frac{99}{200} + \frac{2}{200} = \frac{101}{200} = 50.5\% $$

***

**Part B**: Suppose you toss a Heads.  What is the probability that the coin you chose from the box is double-sided? 

To find this, we can use Baye's theorem. 

Baye's theorem states: 

$$P(B | H) = \frac{P(H | B)}{P(H)}P(B)$$

Let's define P(B) as the probability that the coin is biased. 

We will define P(H) as the probability of Heads.

As such, P(H) (as we had previously defined) is:

$$ P(H) = \frac{101}{200} $$

Similarly, we know that P(B) (the probability that the coin is biased) is fractional: 

$$ P(B) = \frac{1}{100} $$ 

Finally, we know that P(H | B) (the probability that we get heads, given that the coin is biased) is one:

$$P(H | B) = 1$$

Now we have all the pieces to calculate: 


$$P(B | H) = \frac{P(H | B)}{P(H)}P(B)$$

$$P(B | H) = (\frac{1}{\frac{101}{200} })\frac{1}{100} $$

$$ = \frac{200}{101} * \frac{1}{100}$$

$$= \frac{200}{10,100} = \frac{2}{101} = 1.98\%$$

***

**Part C**: Suppose you flip your coin $n$ times and record each outcome.  Obviously, if you flip Tails you know you picked a fair coin. But, if you flipped $n$ Heads in a row you would have a stronger belief that you in fact grabbed the double-headed coin.  What is the smallest value of $n$ Heads in a row that would make you believe that there's a better-than-half chance that the coin you grabbed is double-headed?  

In order to calculate this, we need to calculate $P(B|H^{n})$

How do we do this?! Well, we need to consider the original question, instead, with additional consideration for every consecutive heads. Let's lead into this:

$$ P(H) = \frac{99}{200} + \frac{2}{200} = \frac{101}{200} = 50.5\% $$
$$ P(HH) = \frac{101}{200}*\frac{101}{200} = 25.50\% $$
$$ P(HHH) = \frac{101}{200}*\frac{101}{200}*\frac{101}{200} = 12.87\% $$

So, 

$$ P(H^{n}) = (\frac{101}{200})^n = 50.5^n\% $$

Yay. Now we can plug this back into Baye's theorem (and while we're at it, because I'm too lazy to type out everything again after writing it out on paper, we're going to throw in the law of total probability on the bottom, and use a program to calculate the result).

$$P(B | H^{n}) = \frac{P(H^{n} | B)}{P(B)P(H^{n}\mid B) + P(H^{n}\mid B^{c})P(B^{c})}P(B)$$

Originally I did this very naively and I got 156% probability (because I left P(B) as was) after 7 heads which is...correct...but the probability is wrong. This reconsideration fixes that- as it will never go above 1.

***

```python
for i in range(1,16,2):
    pb = 1/100
    pHb = 1
    pHbc = 1/2
    pbc = 99/100
    
    x = (1)/((pb*pHb)+(pHbc**(i)*pbc))
    r = str(round(x,2))
    
    print("For consecutive head # {0}, we have {1}% certainty".format(i, r))
```

So, **after 6 consecutive heads, if we roll any additional heads, we are certain with at least 50% "certainty"** (whatever that means) that it is a biased/double-sided coin.

In soccer tournaments, [penalty shootouts](https://en.wikipedia.org/wiki/Penalty_shoot-out_(association_football) are used to determine the outcome of the game if the score is tied after regulation play.   In a penalty shootout, each team takes turns attempting 5 shots on the goal from the penalty mark that are defended only by the opposing team's goalkeeper, with the team that makes more successful kicks being declared the winner of the match.  If after 5 shots the score is still tied, the game enters _sudden death_ and the teams continue to take 1 shot each until one team scores and the other does not. 

Suppose that two teams ( Team $A$ and Team $B~$) are still tied at the end of the first round of 5 penalty shots and are entering sudden death.  Each team will now take one shot each in an attempt to win.  Suppose that the result of one shot is independent of the other shots, Team $A$ makes their shot with probability $p$, and Team $B$ makes their shot with probability $q$.  Let $W$ be the event that "Team $A$ wins the game", $E$ be the event that "the game ends after the first round of sudden death penalty shots", and $T$ be the event that "the game is still tied after the first round of sudden death penalty shots". 

**Part A**: The shootout continues until Team $A$ or Team $B$ wins.  What is the sample space $\Omega$ for this experiment?  Represent the outcomes of individual shots by $G$ for a goal and $M$ for a miss and assume that Team $A$ shoots before Team $B$.

We have a few overall outcomes:

$P(W|E) \quad P($~$W|E) \quad P(T|$~$E) \quad P($~$T|E)$

$P(W|E) = {G,M}$

$P($~$W|E) = {M,G}$

$P(T|$~$E) = {M,M,G} \quad or \quad {M,M,M,G}$

$P($~$T|E) = {G,M} \quad or \quad {M,G}$

$$\Omega= ({GM, MG, MMGM, MMMG, MMMMGM, MMMMMG...})$$
This pattern just goes on.
***

**Part B**: Determine $P(W \mid E)$ in terms of $p$ and $q$.

$$P(W\mid E) = \frac{P(W) \cap P(E)}{P(E)}$$

$$P(W \mid E) = \frac{p(1-q)}{p(1-q)+q(1-p)}$$

**Part C**: Explain why events $W$ and $T$ are independent. 

These are separate events because someone tying and someone winning are two seperate, distinct events. You cannot win and have a tie at the same time. (Well...unless you're taking a participation trophy, which we will pretend is not a thing.)

Therefore by the laws of independence:

$$P(W) = P(W \mid T)$$

**Part D**: Use the Law of Total Probability to show that $P(W) = p(1-q) + (2pq - p - q +1) ~P(W \mid T)$. 

$$P(W) = P(W \mid E) P(E) + P(T)P(W \mid T)$$ 

$$P(W \mid E) = \frac{p(1-q)}{p(1-q) + q(1-p)}$$
$$P(T) = (1-q)(1-p) + pq $$

Plugging in values:

$$P(W) = \frac{p(1-q)(p(1-q)+q(1-p))}{p(1-q)(q(1-p))} + ((1-q)(1-p)+pq)P(W \mid T)$$
Multiply out polynomial:
$$P(W) = \frac{p(1-q)(p(1-q)+q(1-p))}{p(1-q)(q(1-p))} + (2pq-p-q+1)P(W \mid T)$$
Cancel terms: 
$$P(W) = \frac{p(1-q)}{1} + (2pq-p-q+1)P(W \mid T)$$

$$P(W) =  p(1-q) + (2pq-p-q+1) P(W \mid T)$$
And we have arrived.

**Part E**: Use the fact from **Part C** that $P(W \mid T) = P(W)$ to determine $P(W)$ in terms of $p$ and $q$. 

$$P(W) = p(1-q)+(2pq-p-q+1)P(W)$$
Move like terms to same side:

$$P(W) - (2pq-p-q+1)P(W) = p(1-q)$$
Pull out common term:

$$P(W)(1- (2pq-p-q+1)) = p(1-q)$$
Divide off excess:

$$P(W) = \frac{p(1-q)}{(1- (2pq-p-q+1)}$$



**Part F**: Explain why the answers to **Part B** and **Part E** are the same. 

In part E we are saying that:
    $$P(W) = \frac{P(W \mid E)}{1-P(T)}$$
    
We are saying the probability of someone winning is equivalent to the probability that someone wins in the first round, divided by the probability that someone does not Tie (I.E, the game ends that turn).




**Part G**: Explain the significance of the solution technique for $P(W)$ outlined in **Part D** and **Part E**.  Why did we go this route instead of computing $P(W)$ directly?  Can we always use such techniques?  

This solution was a very general approach to a problem whose probabilities were not well defined. **This can always be used in a case where we are considering two independent events**, as we were merely utilizing the law of total probabilities, and law of independence. If they were dependent events (not P(W|E) vs P(W|T)), the solution would be very different.

We went this approach because we knew hardly nothing about the probabilities of each distinct event.

In this problem you will write some code to simulate the game of **Miniopoly** (a very rough approximation of [Monopoly](https://en.wikipedia.org/wiki/Monopoly_(game)).  The game consists of a round board with 40 "squares" numbered from $0$ to $40$, where squares $0$ and $40$ are [coincident](https://en.wikipedia.org/wiki/Coincident) and called **GO**.  A player starts on **GO** with \$200 in their pocket.  On each turn the player rolls a pair of dice and then moves a number of spaces equal to the sum of the dice.  Landing on different squares have different results: 

- Squares 2, 7, 17, 22, 33, and 36 cause you to draw a card.  The cards tell you to collect \$50, \$100, or \$200, or pay the bank \$100 or \$150.  

<p> </p>

- Squares 0, 10, and 20 do not trigger any action 

<p> </p>

- Square 30 is **Go To Jail**, which sends the player directly to **Jail** and ends the turn 

<p> </p>

- If a player starts a turn from **Jail** she rolls two dice: 

    - If the player rolls doubles (i.e. the dice have equal values) the player moves from Square 10 a number of squares equal to the sum of the dice
    
    - If the player rolls anything other than doubles she pays the bank \$10 and remains in **Jail** 
    
<p> </p>

- If the player passes or lands on **GO** she collects \$200. 

<p> </p>

- If the player lands on any other square not mentioned above, she pays the bank an amount equal to the square's number (e.g. if she lands on Square 29 she pays the bank \$29). 

<p> </p>

If at any point the player has \$0 or less they go bankrupt and the game ends. 

<br>

**Part A**: Complete the function miniopoly_turn below. The function should simulate one turn of Miniopoly.  It should take as arguments the player's current location on the board as well as the player's current cash reserves, and return the player's new location and cash reserves after the turn has completed. Note that if a player goes bankrupt during the game you should return their cash as 0 instead of a negative value.  

```python
def diceRoll():
    d = np.random.randint(1,7)
    return d

def miniopoly_turn(state, cash):
    #runs each iteration of your turn.
    d = diceRoll() + diceRoll()
    if(state == 30):
        r1 = diceRoll()
        r2 = diceRoll()
        if(r1 == r2):
            state = 10 + (r1 + r2)
        elif(r1 != r2):
            cash = cash - 10
            return 30, cash
    elif(state != 30):
        d = diceRoll() + diceRoll()
        if(state+d >= 40):
            state = (state + d)%40
            cash += 200
        elif(state + d < 40):
            state = state + d
            

    if(state in [2,7,22,33,36]):
        rand = np.random.randint(5)
        cardList = [50,100,200,-100,-150]
        cash += cardList[rand]
        return state, cash
    elif(state in [0,10,20]):
        #Do nothing...
        return state, cash
    elif(state in [30]):
        return state, cash
    else:
        cash -= state
        return state, cash
```

```python
def game(turns, rounds):
    cashList = []
    bankrupt = 0
    jail = False
    previousState = 0
    passedGo = 0
    jailBankrupt = 0
    goBankrupt = 0
    
    for i in range(0, rounds):
        running = True
        cash = 200
        state = 0
        passedGo = 0
        jail = False
        
        while(running == True and cash > 0):
            for i in range(0,turns):
                previousState = state
                state, cash = miniopoly_turn(state,cash)
                if(previousState > state): #If you passed go, your last state will be larger than your current (i.e. you rolled over)
                    passedGo += 1
                if(state == 30):
                    jail = True
                    
                if(cash < 0):
                    bankrupt = bankrupt+1
                    cash = 0
                    if(jail == True): #If you have visited Jail...
                        jailBankrupt += 1
                    if(passedGo >= 2): #If you have passed go more than twice
                        goBankrupt += 1
                    break
            running = False
        cashList.append(cash)
    return cashList, bankrupt, jailBankrupt, passedGo, goBankrupt
        
rounds = 50000
turns = 30
cashList, bankrupt, jailBankrupt, passedGo, goBankrupt = game(turns, rounds)
```
